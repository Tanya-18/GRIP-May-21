```python
x = iris.drop(columns= 'Species')
y = iris['Species']
print(x.shape)
print(y.shape)
```

### Computing k using the Elbow Method

```python
from sklearn.cluster import KMeans
sse = []
for i in range(1,10):
    kmodel = KMeans(n_clusters=i)
    kmodel.fit(x)
    sse.append(kmodel.inertia_)
print("Model ready")    
```

```python
plt.scatter(range(1,10), sse, color="darkblue")
plt.plot(range(1,10), sse, color="purple")
plt.title("Elbow Method", fontsize=18)
plt.xlabel('Number of clusters')
plt.ylabel('Sum of Squares')
plt.show()
```

#### From the above plot we conclude that k = 3

### Applying the K-means Algorithm

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

### Pair Plot

```python
sns.pairplot(data=iris,hue="Species",palette="Set1", height=2)
plt.title("Pair Plot", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.title("K = 3",fontsize=18)
plt.xlabel("SepalLengthCm")
kmeans = KMeans(n_clusters=3)
x["label"] = kmeans.fit_predict(x)
plt.scatter(x.SepalLengthCm[y_kmeans == 0],x.SepalWidthCm[y_kmeans == 0], c='purple',s=100,label='Iris-Setosa')
plt.scatter(x.SepalLengthCm[y_kmeans == 1],x.SepalWidthCm[y_kmeans == 1], c='blue',s=100,label='Iris-Versicolor')
plt.scatter(x.SepalLengthCm[y_kmeans == 2],x.SepalWidthCm[y_kmeans == 2], c='red',s=100,label='Iris-Virginica')
plt.legend()
plt.scatter(kmeans.cluster_centers_[:,0],
            kmeans.cluster_centers_[:,1],
            marker='o',
            s=200,c='yellow') 
```
